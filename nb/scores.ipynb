```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import scipy.stats
import pickle
%matplotlib inline
```

```python
def plotfigs(fnames, labels, _type='ssim'):
    for e, fname in enumerate(fnames):
        if isinstance(_type, list):
            _t = _type[e]
        else:
            _t = _type
        data = pickle.load(open(fname, 'rb'))[_t]
        print(data.shape)
        if len(data.shape) > 2:
            data = np.reshape(data, (data.shape[0]*data.shape[1], data.shape[2]))
        _mean = data.mean(0)        
        plt.errorbar(range(2,data.shape[1]+2), _mean, label=labels[e])

    plt.legend(loc='upper right')
    plt.gca().yaxis.grid(True)
    plt.ylabel("SSIM")
    plt.xlabel("Timesteps")
    plt.axvline(x=10, color='k')
    plt.show()
```

```python
plotfigs(['pretrained.pkl',  '../cimsgens/bairfilter1/stats.pkl', '../cimsgens/pretrainedtrain/stats.pkl',
          '../cimsgens/bairfiltertrain/stats.pkl',
         '../cimsgens/bairfiltertrainposterior/stats.pkl','../cimsgens/pretrainedtrainpost/stats.pkl',
         'newhlm5contlr.pkl', '../hpcgens/newhlm5contlrtrain.pkl'], 
             ['Baseline', 'bairfilter', 'pretrainedtrain',  'bairfiltertrain', 'bairfiltertrainposterior',
             'pretrainedtrainpost', 'hlm', 'hlmtrain'])
```
